```python
from moviepy.video.io.VideoFileClip import VideoFileClip
```

```python
def hms_to_seconds(hms):
    h, m, s = map(int, hms.split(':'))
    return h * 3600 + m * 60 + s
```

```python
input_video_path = 'MARTIN GARRIX LIVE @ SZIGET 2024 [U3shhh922Oo].mp4'
output_video_path = 'trimmed_video_moviepy.mp4'

start_time = "00:30:43"  # Start time in HH:MM:SS format
end_time = "00:31:43"

start_time_seconds = hms_to_seconds(start_time)
end_time_seconds = hms_to_seconds(end_time)


try:
    video = VideoFileClip(input_video_path)
    trimmed_video = video.subclip(start_time_seconds, end_time_seconds)
    trimmed_video.write_videofile(output_video_path, codec="libx264")
    print("Video trimmed and saved successfully using MoviePy.")
except Exception as e:
    print(f"An error occurred: {e}")
```
